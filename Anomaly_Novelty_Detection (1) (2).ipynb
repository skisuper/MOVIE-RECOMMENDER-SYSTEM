**Novelty Detection:**

* ***Novelty detection*** is the mechanism by which an intelligent organism is able to identify an incoming sensory pattern as being hitherto unknown. If the pattern is sufficiently salient or associated with a high positive or strong negative utility, it will be given computational resources for effective future processing.

```
## loading the required packages 

from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs
from numpy import quantile, where, random
import matplotlib.pyplot as plt
```

```
## We'll create a random sample dataset using the make_blob() function

random.seed(1)
x, _ = make_blobs(n_samples=200, centers=1, cluster_std=.3, center_box=(10,10))
```

```
## visualisation of the dataset
plt.scatter(x[:,0], x[:,1])
plt.show()


```

```
## set estimators number and contamination value in arguments. Contamination defines the proportion of outliers in a dataset
lof = LocalOutlierFactor(n_neighbors=20, contamination=.03)
```

```
##fit the model with x dataset and get the prediction data with the fit_predict() method
y_pred = lof.fit_predict(x)

## extract the negative outputs as the outliers.
lofs_index=where(y_pred==-1)
values = x[lofs_index]

## Finally visualize the results in a plot by highlighting the anomalies with a color.

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0],values[:,1], color='r')
plt.show()

model = LocalOutlierFactor(n_neighbors=20) 
print(model)  
model.fit_predict(x) 
 
lof = model.negative_outlier_factor_
thresh = quantile(lof, .03)
print(thresh) 
 
index = where(lof<=thresh)
values = x[index]

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0],values[:,1], color='r')
plt.show()
```
